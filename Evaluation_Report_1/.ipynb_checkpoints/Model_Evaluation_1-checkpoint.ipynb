```python
import os
import pickle
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import warnings
warnings.filterwarnings("ignore")

```

```python
X_train = pd.read_csv("X_train.csv")
X_test  = pd.read_csv("X_test.csv")

y_train = pd.read_csv("Y_train.csv").squeeze()
y_test  = pd.read_csv("Y_test.csv").squeeze()
```

```python
X_train = X_train.loc[:, ~X_train.columns.str.contains("^Unnamed")]
X_test  = X_test.loc[:, ~X_test.columns.str.contains("^Unnamed")]
```

```python
MODEL_DIR = "."

model_files = [
    "best_lasso_model.pkl",
    "decision_tree_model.pkl",
    "ensemble_pipeline.pkl",
    "gradient_boosting.pkl",
    "knn_model.pkl",
    "linear_model.pkl",
    "random_forest_model.pkl",
    "ridge_model.pkl",
    "svm_model.pkl",
    "xgboost_model.pkl"
]
```

```python
def load_model(path):
    try:
        return joblib.load(path)
    except Exception:
        with open(path, "rb") as f:
            return pickle.load(f)
```

```python
def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae  = mean_absolute_error(y_test, y_pred)
    r2   = r2_score(y_test, y_pred)
    
    return y_pred, rmse, mae, r2
```

```python
results = []
predictions = {}

for file in model_files:
    model_name = file.replace(".pkl", "")
    model_path = os.path.join(MODEL_DIR, file)
    
    model = load_model(model_path)
    
    y_pred, rmse, mae, r2 = evaluate_model(model, X_test, y_test)
    
    predictions[model_name] = y_pred
    results.append([model_name, rmse, mae, r2])
```

```python
results_df = pd.DataFrame(
    results,
    columns=["Model", "RMSE", "MAE", "R2"]
).sort_values(by="RMSE")

results_df
```

```python
results_df.to_csv("model_comparison_metrics.csv", index=False)
```

```python
for model_name, y_pred in predictions.items():
    plt.figure(figsize=(6,6))
    plt.scatter(y_test, y_pred, alpha=0.6)
    plt.plot([y_test.min(), y_test.max()],
             [y_test.min(), y_test.max()],
             linestyle="--")
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title(f"Predicted vs Actual - {model_name}")
    plt.show()
```

```python
def make_1d(y):
    """
    Ensure predictions are 1D.
    - If 2D (n_samples, n_outputs), reduce to 1D
    """
    if isinstance(y, pd.Series):
        return y.values
    y = np.asarray(y)
    if y.ndim > 1:
        return y[:, 0]
    return y
```

```python
for model_name, y_pred in predictions.items():
    
    y_true = make_1d(y_test)
    y_pred = make_1d(y_pred)
    
    # Safety check
    if y_true.shape != y_pred.shape:
        print(f" Skipping residual plot for {model_name} (shape mismatch)")
        continue
    
    residuals = y_true - y_pred
    
    plt.figure(figsize=(6,4))
    plt.scatter(y_pred, residuals, alpha=0.6)
    plt.axhline(0, linestyle="--")
    plt.xlabel("Predicted")
    plt.ylabel("Residuals")
    plt.title(f"Residual Plot - {model_name}")
    plt.show()
```

```python
def plot_feature_importance(model, model_name, X):
    if hasattr(model, "feature_importances_"):
        importances = model.feature_importances_
        
        features = X.columns if isinstance(X, pd.DataFrame) else \
                   [f"Feature_{i}" for i in range(len(importances))]
        
        fi = pd.DataFrame({
            "Feature": features,
            "Importance": importances
        }).sort_values(by="Importance", ascending=False).head(15)
        
        plt.figure(figsize=(7,4))
        plt.barh(fi["Feature"], fi["Importance"])
        plt.gca().invert_yaxis()
        plt.title(f"Feature Importance - {model_name}")
        plt.show()
```

```python
for file in model_files:
    model_name = file.replace(".pkl", "")
    model = load_model(file)
    
    plot_feature_importance(model, model_name, X_test)
```
