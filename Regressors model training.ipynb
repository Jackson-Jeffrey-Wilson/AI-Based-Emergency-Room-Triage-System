```python
#Decision tree regressor
from sklearn.tree import DecisionTreeRegressor
import joblib
import pandas as pd

# Load training data
X_train = pd.read_csv(r"C:\Desktop\Maathu\Maathu\CSE\ML\Practicce\BioFusion\Processed data\X_train.csv")
y_train = pd.read_csv(r"C:\Desktop\Maathu\Maathu\CSE\ML\Practicce\BioFusion\Processed data\y_train.csv")
y_train = y_train.values.flatten()

# Train model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

# Save model
joblib.dump(dt_model, r"C:\Users\venus\AI-Based-Emergency-Room-Triage-System\models\decision_tree_model.pkl")
print("Decision Tree Regressor trained and saved.")
```

```python
#Random Forest regressor
from sklearn.ensemble import RandomForestRegressor
import joblib

# Train model
rf_model = RandomForestRegressor(n_estimators=200, random_state=42)
rf_model.fit(X_train, y_train)

# Save model
joblib.dump(rf_model, r"C:\Users\venus\AI-Based-Emergency-Room-Triage-System\models\random_forest_model.pkl")
print("Random Forest Regressor trained and saved.")
```

```python
#Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor
import joblib

# Train model
gb_model = GradientBoostingRegressor(n_estimators=200, learning_rate=0.1, random_state=42)
gb_model.fit(X_train, y_train)

# Save model
joblib.dump(gb_model, r"C:\Users\venus\AI-Based-Emergency-Room-Triage-System\models\gradient_boosting.pkl")
print("Gradient Boosting Regressor trained and saved.")
```

```python
#XGBoost Regressor
from xgboost import XGBRegressor
import joblib

# Train model
xgb_model = XGBRegressor(n_estimators=200, learning_rate=0.1, random_state=42)
xgb_model.fit(X_train, y_train)

# Save model
joblib.dump(xgb_model, r"C:\Users\venus\AI-Based-Emergency-Room-Triage-System\models\xgboost_model.pkl")
print("XGBoost Regressor trained and saved.")
```

```python
#Ensemble regressor
from sklearn.ensemble import VotingRegressor
import joblib

# Create ensemble using RF, GB, XGB
ensemble_model = VotingRegressor(
    estimators=[
        ('rf', rf_model),
        ('gb', gb_model),
        ('xgb', xgb_model)
    ]
)

# Train ensemble
ensemble_model.fit(X_train, y_train)

# Save ensemble
joblib.dump(ensemble_model, r"C:\Users\venus\AI-Based-Emergency-Room-Triage-System\models\ensemble_pipeline.pkl")
print("Ensemble Regressor trained and saved.")
```

```python
#SVR model
from sklearn.svm import SVR
import joblib
import pandas as pd

# Load training data
X_train = pd.read_csv(r"C:\Desktop\Maathu\Maathu\CSE\ML\Practicce\BioFusion\Processed data\X_train.csv")
y_train = pd.read_csv(r"C:\Desktop\Maathu\Maathu\CSE\ML\Practicce\BioFusion\Processed data\y_train.csv")
y_train = y_train.values.flatten()

# Train SVR model
svm_model = SVR(kernel='rbf', C=100, epsilon=0.1)
svm_model.fit(X_train, y_train)

# Save model
joblib.dump(svm_model, r"C:\Users\venus\AI-Based-Emergency-Room-Triage-System\models\svm_model.pkl")
print("SVR model trained and saved.")
```

```python
#KNN regressor model
from sklearn.neighbors import KNeighborsRegressor
import joblib

# Train KNN Regressor
knn_model = KNeighborsRegressor(n_neighbors=5, weights='distance')
knn_model.fit(X_train, y_train)

# Save model
joblib.dump(knn_model, r"C:\Users\venus\AI-Based-Emergency-Room-Triage-System\models\knn_model.pkl")
print("KNN Regressor trained and saved.")
```
